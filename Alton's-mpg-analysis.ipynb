```python
import seaborn as sns
mpg= sns.load_dataset("mpg")
```

```python
mpg.head()
```

```python
sns.relplot(data=mpg, x="mpg", y="horsepower")

#the higher the horsepower the better fuel economy showing as mpg in the chart you burn more gas
with cars vehicles with a lower horsepower.
```

```python
sns.catplot(data=mpg, x="mpg", y="origin", kind="bar", errorbar = None)
#This bar chart shows that Japan has the best fuel economy for their vehicles compared to usa and europe.
# USA vehicles produce the worst fuel economy vehicles which means they burn the most fuel based on mpg.
```

```python
sns.catplot(data=mpg, x="model_year", y="mpg", kind='box')
#overall the fuel economy has become better gaining in mpg over the years 
# in '70 the median was just over 15 mpg, in '77 the mpg median was just under 20 mpg.
# In '82 the median was over 30 mpg. 
```

```python
sns.displot(data=mpg, x="mpg", hue="origin")
#This chrat shows that USA has the largest array of vehicles representing different fuel economies
#Japan shows more vehicles with a fuel economy of 30 mpg and higher which shows they produce more fuel efficient vehicles.
#Most european car have average range fuel economy.
```

```python
sns.catplot(data=mpg, x="cylinders", y="mpg", kind="box")
#This chart shows that higher cylinder vehicles tend to have low fuel economy
#4 and 5 cylinder vehicles have the best overall fuel econonmy when coparing cylinders to mpg.
```

```python
sns.displot(data=mpg, x="mpg", hue="acceleration")
#This chart shows the higher th acceleration rate the bettter the more mpg you can get out of your vehicle.
#considering acceleration impacts the fuel economy the lower the acceleration the lower the fuel economy.
```
